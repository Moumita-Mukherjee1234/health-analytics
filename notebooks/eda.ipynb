```python
import pandas as pd
import matplotlib.pyplot as plt

master = pd.read_csv("/app/data/cleaned/master_table.csv", parse_dates=[
    "appointment_date","treatment_date","bill_date","date_of_birth","registration_date"
])
```

```python
monthly = master.groupby(master['appointment_date'].dt.to_period('M')).size()

plt.figure()
monthly.plot()
plt.title("Monthly Patient Volume")
plt.xlabel("Month")
plt.ylabel("Number of Appointments")
plt.show()
print("Insight to write: Identify peak months and low months.")
```

```python
spec = master['specialization'].value_counts()

plt.figure()
spec.plot(kind='bar')
plt.title("Appointments by Doctor Specialization")
plt.xlabel("Specialization")
plt.ylabel("Appointments")
plt.show()

print("Insight: Which specialization is most demanded.")
```

```python
plt.figure()
master['cost'].hist(bins=20)
plt.title("Treatment Cost Distribution")
plt.xlabel("Cost")
plt.ylabel("Frequency")
plt.show()
print("Insight: Most treatments are low/medium/high cost?")
```

```python
revenue = master.groupby(master['bill_date'].dt.to_period('M'))['amount'].sum()

plt.figure()
revenue.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
print("Insight: Financial growth pattern.")
```

```python
plt.figure()
master['age'].hist(bins=20)
plt.title("Patient Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print("Insight: Which age group visits most.")
```

```python
top_patients = master['patient_id'].value_counts().head(10)

plt.figure()
top_patients.plot(kind='bar')
plt.title("Top 10 Frequent Patients")
plt.xlabel("Patient ID")
plt.ylabel("Visits")
plt.show()
```
